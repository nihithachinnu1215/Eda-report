## APPARENT TEMPERATURE
Apparent temperature is the temperature equivalent perceived by humans, caused by the combined effects of air temperature, relative humidity and wind speed. The measure is most commonly applied to the perceived outdoor temperature.

```python
df['Apparent Temperature (C)'].describe(percentiles=[0.25,0.50,0.75,0.99,1.00])
```

## The above graph shows APPARENT TEMPERATURE.

```python
k=sns.distplot(df['Apparent Temperature (C)'],kde=False)
k.set_title("Apparent Temperature")
plt.rcParams['figure.figsize']=(15,8)
```

## TEMPERATURE
The degree or intensity of heat present in a substance or object, especially as expressed according to a comparative scale and shown by a thermometer or perceived by touch.

```python
df['Temperature (C)'].describe(percentiles=[0.25,0.50,0.75,0.99,1.00])
```

## Temperature and Apparent Temperature
You can see the clear difference between apparent temperature and temperature.The purple colour denotes that both have intersected this means both have same values in that interval.

```python
sns.distplot(df['Apparent Temperature (C)'], hist=True, kde=False, bins=30, color='red', label='Apparent Temp')
sns.distplot(df['Temperature (C)'], hist=True, kde=False, bins=30, color='blue', label='Temperature')
```

Humidity is the amount of water vapour present in air. Water vapour, the gaseous state of water, is generally invisible to the human eye. Humidity indicates the likelihood for precipitation, dew, or fog to be present. 

The amount of water vapour needed to achieve saturation increases as the temperature increases.
